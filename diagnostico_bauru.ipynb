````python
dfBronze ={}
for prefixo in ETIQUETAS:
    URL = f'/content/drive/MyDrive/arquivos-colab-notebooks/bronze/{prefixo}.xlsx'
    dfBronze[prefixo] = pd.read_excel(URL, sheet_name=0,)

dfBronze['CONGRUÊNCIA']
````

````python
df = {}
dataframe= {}

for prefixo in ETIQUETAS:
  df= dfBronze[prefixo]
  categorias = df.loc[:0]
  valores_acumulados = {}
  num_valores = {}
  debug_objeto ={}

  # Percorrer o DataFrame por linha e coluna
  for index, row in df.iterrows():
      for column_name, value in row.items():
          if isinstance(value, (int, float, np.number)):
              if column_name not in valores_acumulados:
                  valores_acumulados[column_name] = 0
                  num_valores[column_name] = 0
              valores_acumulados[column_name] += value
              num_valores[column_name] += 1


  # Calcular a média dos valores acumulados
  for column_name in valores_acumulados:
      valores_acumulados[column_name] /= num_valores[column_name]

  # Criar o novo DataFrame com os valores acumulados
  df_resultado = pd.DataFrame(valores_acumulados, index=[0])


  dataframe[prefixo] = pd.concat([categorias, df_resultado], axis=0)
````

````python
ETIQUETAS = ['EQUIDADE','DIVERSIDADE','MODULARIDADE','TRANSPARÊNCIA','INOVAÇÃO','CONGRUÊNCIA']
dfSilver['TRANSPARÊNCIA']
````

````python
df={}
dfGold={}
for prefixo in ETIQUETAS:
    df={}
    data_set = dfSilver[prefixo]
    categorias = data_set.iloc[0].tolist()
    valores = data_set.iloc[1].tolist()

    # Criar o DataFrame
    df = pd.DataFrame({'Categoria': categorias, 'Valor': valores})

    # Agrupar por categoria e calcular a média

    valores_numericos = pd.to_numeric(df['Valor'], errors='coerce')
    grouped = valores_numericos.groupby(df['Categoria']).mean()
    dfGold[prefixo] = pd.DataFrame(grouped).reset_index()

dfGold[prefixo]
````

````python
dataframe = df

# Extrair as categorias e os valores do DataFrame
categorias = dataframe['Categoria']
valores = dataframe['Valor']

# Criar o gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(categorias, valores)

# Preencher a área sob a curva
ax.fill(categorias, valores, alpha=0.25)

# Personalizar o gráfico


ax.set_title('Gráfico de Radar')
ax.set_ylim([0,5])  # Definir limite superior do eixo y

# Rotacionar os rótulos das categorias
ax.set_xticklabels(categorias, rotation=45)

# Mostrar o gráfico
plt.show()
````
