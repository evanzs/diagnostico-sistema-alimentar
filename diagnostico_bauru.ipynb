````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
````

````python
url_resposta = '/content/drive/MyDrive/arquivos-colab-notebooks/DiagnosticoSistemaAlimentarBauruSP.xlsx'

dfRespostas = pd.read_excel(url_resposta, sheet_name=0)
````

````python
dfRespostas
````

````python
ETIQUETAS = ['EQUIDADE','DIVERSIDADE','MODULARIDADE','TRANSPARÊNCIA','INOVAÇÂO','CONGRUÊNCIA']

dataframes_filtrados = {}
for prefixo in ETIQUETAS:
    colunas_filtradas = dfRespostas.filter(like=prefixo)
    dataframes_filtrados[prefixo] = colunas_filtradas


dataframes_filtrados['EQUIDADE']
````

````python
df = dataframes_filtrados['EQUIDADE']
categorias = df.loc[:0]
valores_acumulados = {}
num_valores = {}

# Percorrer o DataFrame por linha e coluna
for index, row in df.iterrows():
    for column_name, value in row.iteritems():
        if isinstance(value, (int, float, np.number)):
            if column_name not in valores_acumulados:
                valores_acumulados[column_name] = 0
                num_valores[column_name] = 0
            valores_acumulados[column_name] += value
            num_valores[column_name] += 1


# Calcular a média dos valores acumulados
for column_name in valores_acumulados:
    valores_acumulados[column_name] /= num_valores[column_name]

# Criar o novo DataFrame com os valores acumulados
df_resultado = pd.DataFrame(valores_acumulados, index=[0])


dataframe2 = pd.concat([categorias, df_resultado], axis=0)
dataframe2
````

````python
data_set =dataframe2
# Combinar as categorias e valores usando zip()
categorias = data_set.iloc[0].tolist()
valores = data_set.iloc[5].tolist()

# Criar o DataFrame
df = pd.DataFrame({'Categoria': categorias, 'Valor': valores})

# Agrupar por categoria e calcular a média

valores_numericos = pd.to_numeric(df['Valor'], errors='coerce')
grouped = valores_numericos.groupby(df['Categoria']).mean()
df = pd.DataFrame(grouped).reset_index()

plt.figure(figsize=(12, 10))
ax = plt.subplot(111, polar=True)

sns.lineplot(x='Categoria', y='Valor', data=df, sort=False, color='red')

ax.fill_between(df.index, df['Valor'], color='red', alpha=0.25)
ax.set_xticklabels(df['Categoria'], fontsize=15)


plt.ylim(0, 5) 
plt.title('Gráfico de Radar', fontsize=12)
plt.show()

````

````python
data_set = dataframes_filtrados['CONGRUÊNCIA']

categorias = data_set.iloc[2].tolist()
valores =data_set.iloc[3].tolist()  # Valores personalizados

# Plotar o gráfico de radar
angulos = np.linspace(0, 2 * np.pi, len(categorias), endpoint=False).tolist()
angulos += angulos[:1]  # Repetir o primeiro ângulo para fechar o gráfico

valores = np.concatenate((valores, [valores[0]]))  # Repetir o primeiro valor para fechar o gráfico

plt.figure(figsize=(6, 6))
plt.polar(angulos, valores, marker='o')
plt.fill(angulos, valores, alpha=0.3)
plt.xticks(angulos[:-1], categorias)
plt.title('Gráfico de Radar')
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_set = [['Teste 1', 'Teste 2', 'Teste 3'],['pergunta 1','pergunta 2','pergunta 3'], [1, 2, 3]]

df = pd.DataFrame(data_set[2:], columns=data_set[1])

# Plotar o gráfico de radar
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, polar=True)

for i, row in df.iterrows():
    valores = row.values.astype(int)
    categorias = df.columns.tolist()
    angulos = [2 * np.pi * (j / len(categorias)) for j in range(len(categorias))]
    angulos += angulos[:1]
    valores = np.concatenate((valores, [valores[0]]))

    ax.plot(angulos, valores, marker='o')
    ax.fill(angulos, valores, alpha=0.3)

ax.set_xticks(angulos[:-1])
ax.set_xticklabels(categorias)

plt.title('Gráfico de Radar')
plt.show()
````
